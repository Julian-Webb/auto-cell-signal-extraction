```python
import colorsys


def generate_distinct_colors(n):
    # Generate evenly spaced hues in the color spectrum
    hues = [i / n for i in range(n)]

    # Convert hues to RGB
    colors = []
    for hue in hues:
        r, g, b = colorsys.hsv_to_rgb(hue, 0.8, 0.8)  # You can adjust saturation and value as needed
        colors.append((r, g, b))

    return colors
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

colors = generate_distinct_colors(20)

# Create a figure and axis
fig, ax = plt.subplots()

# Create a rectangle patch with width 1 and height 1

rectangle1 = patches.Rectangle((0, 0), 1, 1, linewidth=1, color=colors[0])
rectangle2 = patches.Rectangle((2, 2), 1, 1, linewidth=1, color=colors[1])
rectangle3 = patches.Rectangle((5, 4), 1, 1, linewidth=1, color=colors[2])

# Add the rectangle patch to the axis
ax.add_patch(rectangle1)
ax.add_patch(rectangle2)
ax.add_patch(rectangle3)

# Set the aspect ratio of the plot to be equal, ensuring the square is not distorted
ax.set_aspect('equal')

# Set the limits of the plot
ax.set_xlim(0, 7)
ax.set_ylim(0, 5)

plt.gca().invert_yaxis()

# Show the plot
plt.show()
```

```python
clusters = [[(1, 1), (1, 2), (1, 3), (2, 1)],
            [(0, 0), (0, 1), (0, 2), (0, 3), (1, 0), (2, 0), (2, 2), (2, 3), (3, 0), (3, 1)], [(3, 2)], [(3, 3)]]
roi_clusters_dict = {(0, 0): 1, (0, 1): 1, (0, 2): 1, (0, 3): 1, (1, 0): 1, (1, 1): 0, (1, 2): 0, (1, 3): 0, (2, 0): 1,
                     (2, 1): 0, (2, 2): 1, (2, 3): 1, (3, 0): 1, (3, 1): 1, (3, 2): 2, (3, 3): 3}
n_clusters = 4
n_horizontal = 4
n_vertical = 4
roi_dims = (2048 // 4, 1536 // 4)
img_dims = (2048, 1536)

```

```python
def visualize_clusters(roi_clusters_dict, n_clusters: int, roi_dims: tuple[int, int],
                       img_dims: tuple[int, int]) -> plt.figure:
    # #### generate colors for each cluster #####
    colors = generate_distinct_colors(n_clusters)

    # #### plot each ROI as a rectangle with the color of the cluster #####
    fig, ax = plt.subplots()

    for roi_coordinates, cluster in roi_clusters_dict.items():
        width, height = roi_dims
        rect = patches.Rectangle(roi_coordinates, width, height, color=colors[cluster])

        ax.add_patch(rect)
        continue
        

    # manipulate figure properties
    ax.set_aspect('equal')

    img_width, img_height = img_dims
    ax.set_xlim(0, img_width + 1)
    ax.set_ylim(0, img_height + 1)

    plt.gca().invert_yaxis()

    return fig


visualize_clusters(roi_clusters_dict, n_clusters, roi_dims, img_dims)

```

```python
import matplotlib.pyplot as plt
from matplotlib.lines import Line2D

# Example data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

# Plot data
plt.plot(x, y1, color='blue', label='Line 1')
plt.plot(x, y2, color='red', label='Line 2')

# Define custom legend entries with colors
custom_legend = [
    Line2D([0], [0], color='blue', lw=2),
    Line2D([0], [0], color='red', lw=2)
]

# Set legend with custom legend entries and labels
plt.legend(custom_legend, ['Line 1', 'Line 2'])

plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib.patches import Patch

# Example data
x = [1, 2, 3, 4, 5]
y1 = [1, 4, 9, 16, 25]
y2 = [1, 2, 3, 4, 5]

# Plot data
plt.plot(x, y1, color='blue', label='Line 1')
plt.plot(x, y2, color='red', label='Line 2')

# Define custom legend entries with colored rectangles
custom_legend = [
    Patch(facecolor='blue', edgecolor='black'),
    Patch(facecolor='red', edgecolor='black')
]

# Set legend with custom legend entries and labels
plt.legend(custom_legend, ['Line 1', 'Line 2'])

plt.show()
```
